# 1.Load txt data

```python
txt = open('tvrtketab.txt', encoding='utf-8', errors='ignore').read()
print(txt)
```

## - Removing tabs and storing into a list

```python
data = list(map(lambda x:x.split('\t'), txt.split('\n')))
data
```

```python
df= pd.DataFrame(data[1:], columns=data[0])
```

```python
df.head(10)
```

### Removing whitespace and special characters from text

```python
whiteSpace = string.whitespace
specCharacters = '!#$%&\'()*+:;<=>?[\\]^`{|}~'

tableWhiteSpace = str.maketrans('','', whiteSpace)
tableSpecCharacters = str.maketrans('','',specCharacters)

def cleanText(txt):
    text = str(txt)
    text = text.lower()
    removeWhiteSpace = text.translate(tableWhiteSpace)
    removespecCharacters = removeWhiteSpace.translate(tableSpecCharacters)
    return str(removespecCharacters)
    
    
```

```python
df['company_data'] = df['company_data'].apply(cleanText)
```

```python
dataClean = df.query("company_data != '' ")
dataClean.dropna(inplace=True)
```

```python
dataClean.head(10)
```

### Converting clean data into Spacy format

```python
group = dataClean.groupby(by='\ufeffcompany_name')
```

```python
group.groups.keys()
```

### Extracting all data except with tag 'O' from single company

```python
groupArray = group.get_group('genox.JPG')[['company_data','tag']].values
content = ''
annotations = {'entities':[]}
start = 0
end = 0

for text, label in groupArray:
    text = str(text)
    stringLength = len(text) + 1
    start = end
    end = start + stringLength
    
    if label != 'O':
        annot = (start, end - 1, label)
        annotations['entities'].append(annot)
        
    content = content + text + ' '
```

### Looping data from all companies

```python
names = group.groups.keys()
allCompaniesData = [] # empty list for all companies
for name in names:
    companyData = [] # empty list for single company
    groupArray = group.get_group(name)[['company_data','tag']].values #all text data from tag 'company_data'
    content = ''
    annotations = {'entities':[]}
    start = 0
    end = 0

    for text, label in groupArray: #looping thru all'company_data' and 'tag' of single company
        text = str(text)
        stringLength = len(text) + 1
        start = end
        end = start + stringLength

        if label != 'O': #all tags except 'O'
            annot = (start, end - 1, label)
            annotations['entities'].append(annot)

        content = content + text + ' '
    companyData = (content, annotations) #single company with usefull tags
    allCompaniesData.append(companyData) #adding single company into list of companies
```

### Shuffle and split into training / testing set

```python
import random
import pickle
```

```python
random.shuffle(allCompaniesData)
```

```python
TrainData = allCompaniesData[:360] #90% of data
TestData = allCompaniesData[360:] #10% of data
```

```python
with open('data/TrainData.pickle', 'wb') as p:
    pickle.dump(TrainData, p)
with open('data/TestData.pickle', 'wb') as p:
    pickle.dump(TestData, p)
```
